# Iris Flower Classification ML Project
In this project, there are three species of Iris flower. When a new flower is given, it find the class of the flower whether it setosa, versicolor or virginica.

# Importing Important modules
- Import Library
- load dataset
- print head of dataset
- get details of dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
sbn.set(style="white", color_codes=True)
```

```python
iris = pd.read_csv("./DATA/iris.data")
iris.head()
```

```python
iris['Species'].value_counts()
```

So, we have 3 classes Iris-setosa, Iris-versicolor, Iris-virginica each have 50 entries. We have 150 samples in total.

# Visualize the dataset
Here we are visualizing the dataset using Scatter Plot

```python
sbn.FacetGrid(iris, hue="Species", height=6).map(plt.scatter, "Petal Length", "Petal Width").add_legend()
```

# Preparing Input and Output

```python
X = iris[['Sepal Length','Sepal Width', 'Petal Length', 'Petal Width']].values
y = iris[['Species']].values
```

# Generating data model
Here we train the data set and generate a data model Using Logistic Regression Algorithm of ML  to predict the Class / Species of Flower based on given input.

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X,y)
```

Measuring the Accuracy of the data model

```python
model.score(X, y)
```

summarize the data model

```python
from sklearn import metrics
predicted = model.predict(X)
print(metrics.classification_report(predicted, y))
```
